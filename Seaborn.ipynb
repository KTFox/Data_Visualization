```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Set style for seaborn
sns.set_theme()
```

```python
#Available dataset in seaborn
print(sns.get_dataset_names())
```

```python
#Prepare data
df = sns.load_dataset("tips")
df.head()
```

#  Distribution plot

```python
sns.displot(data=df, x="total_bill", col="time", row="smoker", kde=True)
```

#  Bar plot

```python
sns.barplot(data=df, x="sex", y="total_bill", estimator=np.mean);
```

#  Count plot

```python
sns.countplot(data=df, x="smoker");
```

#  Box plot

```python
sns.boxplot(data=df, x="day", y="total_bill", hue="smoker", palette="Blues")
plt.legend(loc=0);
```

# Violin plot

```python
sns.violinplot(data=df, x="day", y="total_bill", hue="sex")
```

# Swarm plot

```python
sns.swarmplot(data=df, x="day", y="total_bill", hue="sex")
```

# Stripp plot

```python
sns.stripplot(data=df, x="day", y="total_bill")
```

#  FacetGrid

```python
fg_tip = sns.FacetGrid(data=df, col="day", row="smoker", hue="sex", palette="Set2", height=4, aspect=1.5)
fg_tip.map(sns.scatterplot, "total_bill", "tip", s=300, edgecolor="black", alpha=0.5).add_legend()
```

#  Joint plot

```python
sns.jointplot(data=df, x="total_bill", y="tip")
```

```python
 sns.jointplot(data=df, x="total_bill", y="tip", hue="sex")
```

#  Pair plot

```python
sns.pairplot(data=df, hue="sex", height=4)
```
